```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Carbon_emission_data.csv")
```

```python
df.head()
```

```python
df2 = df.groupby('Country')['Carbon_Emission_(kg/km)'].sum().reset_index()
print(df2)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
```

```python
df2.plot(kind='bar', x='Country', y='Carbon_Emission_(kg/km)', ax=ax)
plt.show()
```

```python
ax = plt.plot(df2['Country'], df2['Carbon_Emission_(kg/km)'])
plt.xlabel('Country')
plt.ylabel('Carbon Emission (kg/km)')
plt.title('Carbon Emission by Country')
plt.xticks()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df_final = pd.merge(df, df_b, on='Country', how='inner')

print(df_final.head())
```

```python
yearly_emission_bycountry = df_final.groupby('Country')['Carbon_Emission_(kg/km)'].sum().reset_index()
yearly_emission_bycountry.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(yearly_emission_bycountry['Country'], yearly_emission_bycountry['Carbon Emission (kg/km)'])
plt.xlabel('Country')
plt.ylabel('Total Carbon Emission (kg/km)')
plt.title('Yearly Total Carbon Emission by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
